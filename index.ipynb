## Create the word frequency dictionary for each class

Create a word frequency dictionary for each class.

```python
freq_dict = {}
for target in ["ham", "spam"]:
    target_dict = {}
    for text in list(train_df.loc[train_df.label==target, "text"]):
        for word in text.split():
            if word in target_dict.keys():
                target_dict[word] += 1
            else:
                target_dict[word] =1
    freq_dict[target] = target_dict
```
